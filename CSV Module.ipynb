```python
import csv
```

```python
open_file = open('student_data.csv', mode='r')
open_file
```

```python
reader = csv.reader(open_file)
reader
```

```python
data = list(reader)
data
```

```python
data
```

```python
header = data[0]
header
```

```python
header.append('Percentage')
header
```

```python
grades = []

# Step 3: Loop through each row (skip header row)
for row in data[1:]:
    percent = float(row[5])  # Assuming 6th column is Percentage
    if percent >= 85:
        grade = "A"
    elif percent >= 70:
        grade = "B"
    elif percent >= 55:
        grade = "C"
    elif percent >= 40:
        grade = "D"
    else:
        grade = "F"
    grades.append(grade)

# Step 4: Add "Grade" column header
data[0].append("Grade")

# Step 5: Add grade values to each row
for i in range(1, len(data)):
    data[i].append(grades[i - 1])

# Step 6: Print updated data
for row in data:
    print(row)
```
