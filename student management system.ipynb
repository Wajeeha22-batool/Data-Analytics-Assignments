***Student Management System***

Class: Student

Attributes: name, roll_number, grades

Method: calculate_average()

Class: Course

Holds multiple students and can calculate class average.

```python
class Student:
  def __init__(self,name,roll_number,grades):
    self.name = name
    self.roll_number = roll_number
    self.grades = grades
  def calculate_average(self):
    return sum(self.grades)/len(self.grades)

# Creating objects:
student1 = Student("wajeeha", "H2006075",[78,50,34,26,90])
student2 = Student("Ali", "H2006076",[81,78,98,56,43])

# Printing the result of object:
print(f"{student1.name} having roll number {student1.roll_number} --> Average: {student1.calculate_average() :.2f}")
print(f"{student2.name} having roll number {student2.roll_number} --> Average: {student2.calculate_average() :.2f}")
```
