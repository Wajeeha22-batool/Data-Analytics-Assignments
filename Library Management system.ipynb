Library Management System

Create a Book class with attributes like title, author, year.

A Library class can store multiple books and allow:

Adding books

Searching books

Borrowing/returning books

```python
  class Book:
    def __init__(self,title,author,year):
      self.title = title
      self.author = author
      self.year = year
      self.is_borrowed = False
  class Library:
    def __init__(self):
      self.books = []
    def add_books(self,book):
      self.books.append(book)
      print(f"book added:", {book.title})
    def display_books(self):
      for book in self.books:
        print(f"The {book.title} by {book.author} in {book.year}")
    def search_book(self,title):
      for book in self.books:
        if book.title ==title:
          return book
        return None
    def borrow_book(self,title):
      book = self.search_book(title)
      if book:
        if not book.is_borrowed:
          book.is_borrowed = True
          print(f"You borrowed: {book.title}")
        else:
          print(f" {book.title} is already borrowed")
      else:
        print("book not found")
    def return_books(self,title):
      book = self.search_book(title) # Define book here
      if book:
        if book.is_borrowed:
          book.is_borrowed = False
          print(f"you returned the book: {book.title}")
        else:
          print(f"you have not borrowed the book: {book.title}")
      else:
        print("book not found")

  # Adding books to library:
  library = Library()
  library.add_books(Book("The Great Gatsby","F. Scott Fitzgerald",1925))
  library.add_books(Book("To Kill a Mockingbird","Harper Lee",1960))
  library.add_books(Book("Silent patient","George Orwell",1949))
```

```python
library.display_books()
```

```python
library.borrow_book("Silent patient")
```

```python
library.search_book("The Great Gatsby")
```

```python
library.borrow_book("The Great Gatsby")
```

```python
library.return_books("The Great Gatsby")
```

```python
library.return_books("Silent patient")
```

```python
library.return_books("To Kill a Mockingbird")
```
