```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://countrycode.org/"
url
```

```python
page = requests.get(url)
page
```

```python
soup = BeautifulSoup(page.text, 'html')
soup
```

```python
print(soup.prettify())
```

```python
soup.find('div')
```

```python
soup.find_all('div')
```

```python
soup.find('p')
```

```python
soup.find('p').text.strip
```

```python
header = soup.find_all('th')
header
```

```python
title = [i.text.strip() for i in header]
title
```

```python
table= soup.find('table')
table
```

```python
rows = []

for tr in table.find_all("tr")[1:]:  # skip header row
    row = [td.text.strip() for td in tr.find_all("td")]

    # FIX: make row length equal to header length
    if len(row) < len(title):
        row += [''] * (len(title) - len(row)) 
    rows.append(row)   
rows
```

```python
df = pd.DataFrame(rows, columns=title)
df
```

```python
df.to_csv("country_codes.csv", index=False)
```
