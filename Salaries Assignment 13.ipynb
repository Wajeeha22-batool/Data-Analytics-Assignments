```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Salaries.csv", low_memory=False)
```

```python
df.columns
```

Drop rows where EmployeeName and JobTitle are 'Not provided'

```python
df.head(5)
```

```python
df.drop(df[(df["EmployeeName"] == "Not provided") & (df["JobTitle"] == "Not provided")].index, inplace=True)
df
```

Replace 'Not Provided' with 0 in BasePay OvertimePay	OtherPay	Benefits

```python
cols = ["BasePay", "OvertimePay", "OtherPay", "Benefits"]
df[cols] = df[cols].replace('Not Provided', 0)
df[cols]
```

Change BasePay	OvertimePay	OtherPay	Benefits type to float

```python
cols = ["BasePay", "OvertimePay", "OtherPay", "Benefits"]
df[cols] = df[cols].astype(float)
df[cols]
```

Display all employees name and basepay columns who have a BasePay greater than 200,000.

```python
# df.columns = df.columns.str.strip()
df.head(5)
```

```python
# Convert BasePay to numeric to handle mixed data types
df["BasePay"] = pd.to_numeric(df["BasePay"], errors="coerce")
df[df["BasePay"] > 200000][["EmployeeName", "BasePay"]]
```

Find all employees whose JobTitle contains the word “Manager” (case = False)

```python
df[df.JobTitle.str.contains('Manager', case=False)]
```

Retrieve only the columns: EmployeeName, JobTitle, and TotalPay.

Find count of employees who have zero OvertimePay.

Calculate the TotalPay for each JobTitle where totalpay > 300000

```python
df[df["TotalPay"] > 300000].groupby("JobTitle")["TotalPay"].sum().reset_index()
```

Find the top 5 JobTitles with the highest average TotalPayBenefits show in descending order

```python
df.groupby("JobTitle")["TotalPayBenefits"].mean().sort_values(ascending=False).head(5)
```

Replace all missing Benefits values with 0

```python
df['Benefits'].fillna(0)
```

Identify the top employee name by TotalPay in each Year. Show year, employeename and totalpay dataframe

```python
# Ensure TotalPay is numeric
df["TotalPay"] = pd.to_numeric(df["TotalPay"], errors="coerce").fillna(0)
top_employee_each_year = df.loc[df.groupby("Year")["TotalPay"].idxmax(), ["Year", "EmployeeName", "TotalPay"]]
print(top_employee_each_year)
```
