#### Find upper bound, lower bound, number of outliers in Engine HP

```python
# df['Engine HP'].dropna()
Q1 = df["Engine HP"].quantile(0.25)
Q3 = df["Engine HP"].quantile(0.75)
IQR = Q3 - Q1
print(Q1)
print(Q3)
print(IQR)
```

```python
# Calculate bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
print(lower_bound)
print(upper_bound)
```

```python
df_filtered = df[(df["Engine HP"] >= lower_bound) | (df["Engine HP"] <= upper_bound)]

print(f"Original Rows: {df.shape[0]}, Filtered Rows: {df_filtered.shape[0]}")
```

#### Fill missing values in Number of Doors column with median value then check if values are filled
