```python
import pandas as pd
import sqlite3 as sql
```

```python
conn = sql.connect('database.sqlite')
conn
```

```python
pd.read_sql_query('select * from Salaries', conn)
```

#### **1. What is the average TotalPayBenefits for each JobTitle? exclude ('Not Provided', 'Not provided')**

```python
pd.read_sql_query("SELECT JobTitle,AVG(TotalPayBenefits) AS Avg_TotalPayBenefits FROM salaries WHERE JobTitle NOT IN ('Not Provided', 'Not provided') GROUP BY JobTitle ORDER BY Avg_TotalPayBenefits DESC;", conn)
```

#### **2. Show The JobTitle and Overtime Pay for All Employees With Overtime Pay greater than 50000**

```python
pd.read_sql_query("SELECT id, EmployeeName, JobTitle, OvertimePay FROM salaries where OvertimePay>50000", conn)
```

#### **3. What is the total number of unique employees? exclude ('Not Provided', 'Not provided')**

```python
pd.read_sql_query("SELECT  COUNT(DISTINCT EmployeeName) AS Unique_Employees FROM salaries WHERE EmployeeName NOT IN ('Not Provided', 'Not provided')",conn)
```

#### **4.Show The Top 10 highest Paid Employees**

```python
pd.read_sql_query("select EmployeeName, JobTitle, TotalPayBenefits from salaries order by TotalPayBenefits DESC limit 10",conn)
```

#### **5.Show All Employees Ordered By Their TotalPay Benefits In Descending Order**

```python
SELECT EmployeeName, JobTitle, TotalPayBenefits
FROM salaries
ORDER BY TotalPayBenefits DESC;
```
