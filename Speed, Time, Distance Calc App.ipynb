Write a Python program to calculate Speed, Distance, or Time based on user input.

### Program Title:

Speed, Distance, and Time Calculator

### Program Objective:

To develop a program that calculates Speed, Distance, or Time based on the user’s choice and input values, using the standard motion formulas.

### Introduction:

This program helps users calculate one of the three motion parameters — Speed, Distance, or Time — when the other two are known.

The formulas used are:

- Speed = Distance / Time

- Distance = Speed × Time

- Time = Distance / Speed

The program provides a menu-driven system where the user can choose what to calculate. It also includes error handling for invalid inputs, negative values, and division by zero.

#### Program Logic:

Start the Program

Display Menu Options:

1. Calculate Speed
2. Calculate Distance
3. Calculate Time
4. Exit


Ask the user to choose an option (1–4).

Based on the user’s choice:

If choice = 1 (Speed):

Input: Distance, Time

Check if inputs are positive and Time ≠ 0

Formula: Speed = Distance / Time

Display Speed

If choice = 2 (Distance):

Input: Speed, Time

Check if inputs are positive

Formula: Distance = Speed × Time

Display Distance

If choice = 3 (Time):

Input: Distance, Speed

Check if inputs are positive and Speed ≠ 0

Formula: Time = Distance / Speed

Display Time

If choice = 4 (Exit):

Display a goodbye message

Stop the program

If the user enters an invalid choice:

Display an error message

Show the menu again

Repeat the process until the user selects "Exit".

```python
# Program Title: Speed, Distance, and Time Calculator

print("Speed, Distance, and Time Calculator")

# Program runs until the user chooses to exit
for i in range(100):  # loop to repeat menu (you can use while True instead)
    print("\nMenu:")
    print("1. Calculate Speed")
    print("2. Calculate Distance")
    print("3. Calculate Time")
    print("4. Exit")

    choice = input("Enter your choice (1-4): ")

    # Calculate Speed
    if choice == '1':
        try:
            distance = float(input("Enter Distance (in km): "))
            time = float(input("Enter Time (in hours): "))
            if distance > 0 and time > 0:
                speed = distance / time
                print(f"Speed = {speed:.2f} km/h")
            else:
                print("Error: Distance and Time must be positive numbers.")
        except:
            print("Invalid input! Please enter numeric values.")
        

    # Calculate Distance
    elif choice == '2':
        try:
            speed = float(input("Enter Speed (in km/h): "))
            time = float(input("Enter Time (in hours): "))
            if speed > 0 and time > 0:
                distance = speed * time
                print(f"Distance = {distance:.2f} km")
            else:
                print("Error: Speed and Time must be positive numbers.")
        except ValueError:
            print("Invalid input! Please enter numeric values.")

    # Calculate Time
    elif choice == '3':
        try:
            distance = float(input("Enter Distance (in km): "))
            speed = float(input("Enter Speed (in km/h): "))
            if distance > 0 and speed > 0:
                time = distance / speed
                print(f"Time = {time:.2f} hours")
            else:
                print("Error: Distance and Speed must be positive numbers.")
        except:
            print("Invalid input! Please enter numeric values.")
        

    # Exit program
    elif choice == '4':
        print("Thank you for using the Speed, Distance, and Time Calculator. Goodbye!")
        break

    # Invalid choice
    else:
        print("Invalid choice! Please select a valid option (1–4).")
```
